# Heuristics

```python

from models.path.ActionPathHeuristic import ActionPathHeuristic, ActionPathHeuristics


heuristic = ActionPathHeuristic(
    blockMap, 
    coordinateMultiplier=2,
    heuristic=ActionPathHeuristics.GoalDistanceFlood
)

print(heuristic.getGoalDistance(blockMap.startPosition))
```

```python
plot = PlotTools.create(figsize=60)


plot.plotTensor(tensor=heuristic.getDistanceTensor(), colors=['red', 'blue'], plotValues=False, extent=Rect(blockMap.rtree.bounds))
plot.plotBlocks(blockMap.blocks, alpha=1)
```
