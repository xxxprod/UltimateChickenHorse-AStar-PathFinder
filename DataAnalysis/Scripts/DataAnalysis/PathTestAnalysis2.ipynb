```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from services.UCHPathFinder import *
```

```python
class TestStatisticsDecoder(json.JSONDecoder):
    def decode(self, jsonStr):
        data = super().decode(jsonStr)
        for testScenario in data:
            for row in testScenario[1:]:
                row['path'] = ActionPathDecoder().deserializePath(row['path'])
        return data
```

```python
def loadBestResults(levelName):

    path = LevelPathBuilder.getLevelDirectory(levelName)+'/01_TestStatistics.json'

    bestPaths = []
    with open(path) as json_file:
        results = json.load(json_file, cls=TestStatisticsDecoder)
        for result in results:
            if len(result) < 2:
                continue
            bestPath:dict = sorted(result[1:], key=lambda x: len(x['path']), reverse=False)[0]
            bestPath.update(result[0])
            bestPaths.append(bestPath)

    if len(bestPaths) == 0:
        return None

    return sorted(bestPaths, key=lambda x: len(x['path']), reverse=False)[0]
```

```python
def executeBestPaths(levelName):
    bestResult = loadBestResults(levelName)

    if bestResult is None:
        print('No best result found')
        return

    print(bestResult)

    path = bestResult['path']

    pathFinder = UCHPathFinder(
        levelName, 
        # quantizationFactor=(3, 3), 
        # costWeights=[1],
        forceLoadLevel=False, 
        forceCreateNodeMap=False
    )

    pathFinder.executePath(path, shutdownServer=False)

    return bestResult
    # break
```
