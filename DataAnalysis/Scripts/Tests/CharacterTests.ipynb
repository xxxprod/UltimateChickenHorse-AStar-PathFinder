```python
from utils.tools import *

import matplotlib as mpl
import numpy as np


colors = [
    'red',
    'blue',
    'green',
    'yellow',
    'orange',
    'purple',
    'pink',
    'brown',
    'gray',
    'cyan',
    'magenta',
    'lime',
    'olive',
    'teal',
    'navy',
]

globalScale = character['scale']
globalOffset = mainCollider['offset']
globalOffset = [globalOffset[0], globalOffset[1] - 0.25]
# globalOffset = [0, 0]

plot = PlotTools.create(facecolor='white', figsize=(6,6))
# plot.axis[0].grid()
# plot.axis[0].set_xbound(-2, 2)
# plot.axis[0].set_ybound(-3, 4)
plot.axis[0].set_xlim(-1, 3)
plot.axis[0].set_ylim(-0.5, 3)
# plot.axis[0].set_xticks(np.arange(-1, 1, 0.5))


mainOffset = mainCollider['offset']
size = mainCollider['size']
x = (mainOffset[0] - size[0] / 2 + globalOffset[0]) * globalScale[0]
y = (mainOffset[1] - size[1] / 2 + globalOffset[1]) * globalScale[1]            
w = size[0] * globalScale[0]
h = size[1] * globalScale[1]

print (mainCollider['name'], round(x, 2), round(y, 2), round(x+w, 2), round(y+h, 2))
rect = [x, y, x + w, y + h]
plot.plotRectangle(rect, edgeColor='#00aaff', label=mainCollider['name'])

i = 0
for obj in byName.keys():

    for collider in byName[obj]:

        i += 1

        offset = collider['offset']
        localScale = collider['scale']
        localPos = collider['localPos']
        localScale = [localScale[0] * globalScale[0], localScale[1] * globalScale[1]]


        if collider['type'] == 'BoxCollider' or collider['type'] == 'CapsuleCollider':
            size = collider['size']
            x = (offset[0] - size[0] / 2 + localPos[0] + globalOffset[0]) * localScale[0]
            y = (offset[1] - size[1] / 2 + localPos[1] + globalOffset[1]) * localScale[1]            
            w = size[0] * localScale[0]
            h = size[1] * localScale[1]

            color = colors[i % len(colors)]
            print (collider['name'], round(x, 2), round(y, 2), round(x+w, 2), round(y+h, 2), color)
            rect = [x, y, x + w, y + h]
            label = collider['name'] #+ '_' + collider['type']
            plot.plotRectangle(rect, color=color, label=label, alpha=0.5)

        elif collider['type'] == 'CircleCollider':
            radius = collider['radius'] * localScale[0]
            x = offset[0] + localPos[0]
            y = offset[1] + localPos[1]

            color = colors[i % len(colors)]
            print (collider['name'], round(x, 2), round(y, 2), round(x+w, 2), round(y+h, 2), color)
            rect = [x - radius, y - radius, x + radius, y + radius]
            label = collider['name'] #+ '_' + collider['type']
            plot.plotRectangle(rect, color=color, label=label, alpha=0.5)

    plot.axis[0].legend( bbox_to_anchor=(0.55, 0.8), loc='upper left', borderaxespad=0. )

def add_interval(ax, xdata, ydata, caps="  "):
    line = ax.add_line(mpl.lines.Line2D(xdata, ydata, color='black'))
    anno_args = {
        'ha': 'center',
        'va': 'center',
        'size': 10,
        'color': 'black'
    }
    distance = xdata[1] - xdata[0]
    xPos = xdata[0] + distance / 2 + 0.0
    a0 = ax.annotate(caps[0], xy=(xdata[0]-0.01, ydata[0]), **anno_args)
    a1 = ax.annotate(caps[1], xy=(xdata[1], ydata[1]), **anno_args)
    ax.annotate(round(distance, 2), xy=(xPos, ydata[0] + 0.06), **anno_args)
    return (line,(a0,a1))

def add_interval90(ax, xdata, ydata, caps="  "):
    line = ax.add_line(mpl.lines.Line2D(xdata, ydata, color='black'))
    anno_args = {
        'ha': 'center',
        'va': 'center',
        'size': 10,
        'color': 'black',
        'rotation': 90
    }
    ax.annotate(caps[0], xy=(xdata[0], ydata[0]+0.01), **anno_args)
    ax.annotate(caps[1], xy=(xdata[1], ydata[1]+0.02), **anno_args)
    distance = ydata[1] - ydata[0]
    yPos = ydata[0] + distance / 2 + 0.05
    ax.annotate(round(distance, 2), xy=(xdata[0]-0.07, yPos), **anno_args)
    # return (line,(a0,a1))

def add_interval270(ax, xdata, ydata, caps="  "):
    line = ax.add_line(mpl.lines.Line2D(xdata, ydata, color='black'))
    anno_args = {
        'ha': 'center',
        'va': 'center',
        'size': 10,
        'color': 'black',
        'rotation': -90
    }
    ax.annotate(caps[0], xy=(xdata[0], ydata[0]+0.00), **anno_args)
    ax.annotate(caps[1], xy=(xdata[1], ydata[1]+0.01), **anno_args)
    distance = ydata[1] - ydata[0]
    yPos = ydata[0] + distance / 2 + 0.05
    ax.annotate(round(distance, 2), xy=(xdata[0]+0.04, yPos), **anno_args)
    # return (line,(a0,a1))


print([c['size'][0] for c in colliders if str(c['name']).startswith('UpperBody')])
print([c['size'][0] for c in colliders if str(c['name']).startswith('LowerBody')])

add_interval90(plot.axis[0], [-0.75, -0.75], [0, 1.95], ['|', '|'])
add_interval270(plot.axis[0], [0.75, 0.75], [0, 1.69], ['|', '|'])
add_interval(plot.axis[0], [-0.45, 0.45], [2.45, 2.45], ['|', '|'])
add_interval(plot.axis[0], [-0.39, 0.39], [2.25, 2.25], ['|', '|'])
add_interval(plot.axis[0], [-0.33, 0.3], [2.05, 2.05], ['|', '|'])

plot.fig.savefig(f'{dataAnalysisRoot}/TestResultPlots/CharacterHitbox.png', dpi=300, bbox_inches='tight')
```
